## Exercise 4

- Write a program that uses a `for` loop to read in 5 numbers and output the total.

````python
total = 0

for i in range(5):
  while True:
    num = input(f"Enter number {i+1} : ")

    try:
      num = int(num)
      break

    except ValueError:
      print(f"You entered {num} as input but program only accept integers! Please enter an integer!")
  total += num

print(f"Total is {total}")
````

## Exercise 6 - Display stars

- Write a program which ask the user how many stars(*) are required and then make it display that number of stars across the screen.

````python
while True:
  num = input("Enter number : ")

  try:
    num = int(num)
    break

  except ValueError:
    print("Enter an integer!")

for i in range(num+1):
  print(" " * (num-i) + "* " * (i))
````

## Exercise 7 - Rolling two dice & counting double generated
- Write a program that simulates the rolling of two six-sided dice and then counts the number of times that dice generate a double(e.g.: Both dice show same value). Roll the dice 100 times and then print the number of doubles. **Output:** "Out of 100 you rolled 20 doubles."

````python
import random

count = 0

for i in range(100):
  first_dice = random.randint(0,6)
  second_dice = random.randint(0,6)

  if( first_dice == second_dice):
    count += 1

print(f"Out of 100 you rolled {count} doubles.")
````
