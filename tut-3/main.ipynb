The program should do the following.
- Ask the user to enter "1" if they want to convert from Celsius to Fahrenheit or "2" if they want to convert from Fahrenheit to Celsius.
- Ask the user to enter a temperature in correct unit.

Formulas:
    f = (c * 1.8) = 32
    c = (f - 32) / 1.8

- Apply the conversion formula and display the converted temperature.
- Use an if or check if the user has enterd "1", elif to check if the user has entered "2" and an else to display "Invalid option entered!"

```python
f = 0
c = 0 
unit = 0

print("Enter your temperature unit : \n (1) - Celsius to Fahrenheit \n (2) - Fahrenheit to celsius")

unit = int(input("Enter the number : "))

if (unit == 1):
    c = int(input("Enter celsius value : "))
    f = (c * 1.8) + 32
    print("Fahrenheit value is", f)

elif (unit == 2):
    f = int(input("Enter Fahrenheit value : "))
    c = (f - 32) / 1.8
    print("Celsius value is", c)

else: 
    print("invalid option entered.")
```
